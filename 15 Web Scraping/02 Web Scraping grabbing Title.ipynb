```python
base_url = 'http://books.toscrape.com/catalogue/page-{}.html'
two_star_titles = []

# since there are 50 pages
for n in range(1,51):
    scrape_url = base_url.format(n)
    req = requests.get(scrape_url)
    soup = bs4.BeautifulSoup(req.text,'lxml')
    products = soup.select('.product_pod')    
    # 20 books on every single page
    for i in range(0,20):
        example = products[i]
        if [] != example.select('.star-rating.Two'):
            title = example.select('a')[1]['title']
            two_star_titles.append(title)
```

```python
# GOAL ACCOMPLISHED
for i,n in enumerate(two_star_titles):
    print(i+1,n)
```

```python
# BETTER LOGIC

base_url = 'http://books.toscrape.com/catalogue/page-{}.html'
two_star_titles = []

for n in range(1,51):
    scrape_url = base_url.format(n)
    req = requests.get(scrape_url)
    soup = bs4.BeautifulSoup(req.text,'lxml')
    books = soup.select('.product_pod')
    
    for book in books:
        if len(book.select('.star-rating.Two')) != 0:
            book_title = book.select('a')[1]['title']    
            two_star_titles.append(book_title)
```
