## Tic Tac Toe

```python
from IPython.display import clear_output
from random import randint

def display_board(board):
    '''Display the game board for Visualization.'''
    clear_output()  
    # Above line will clear the previous output
    # Remember, this only works in jupyter nb!
    print('   |   |')
    print(' ' + board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |')
    

def player_input():
    '''Returns the tuple of player1 and player2 marker.'''
    marker = ''
    # KEEP ASKING PLAYER 1 TO CHOOSE X OR O 
    while marker != 'X' and marker != 'O':
        marker = input("For Player 1 choose 'X' or 'O': ").upper()     
    if marker == 'X':
        return ('X', 'O')
    else:
        return ('O', 'X')
    
    
def choose_first():
    '''Choose randomly which player will play first.'''
    flip = randint(0,1)
    if flip == 0:
        return "Player 1"
    else:
        return "Player 2"

    
def space_check(board,postion):
    '''checks whether board's position is empty or not, if empty: "True", else: "False"'''
    return board[postion]==' '
 
    
def full_board_check(board):
    '''check whether board is completely filled or not, if completely filled: "True", else: "False"'''
    for i in range(1,10):
        if space_check(board,i):
            return False
    return True

    
def player_choice(board):
    '''players choice for entering position on board'''
    # initialize position to 0 to enter the loop at least once.
    position = 0
    # check if the entered position is between 1 and 9 & also
    # check if the position is available on the board.
    while position not in [1,2,3,4,5,6,7,8,9] or not space_check(board,position):
        try:
            position = int(input("Enter the position: (1-9)"))
        except:
            print("Please enter a number! ")
    return position


def place_marker(board,marker,position):
    '''places player's marker at that position on board'''
    board[position]=marker


def win_check(board, marker):
    '''Checks game is whether won by player or not'''
    # All Rows, and check to see if they share the same marker
    return ((board[1]==marker and board[2]==marker and board[3]==marker) or
    (board[4] == board[5]== board[6] == marker) or
    (board[7] == board[8]== board[9] == marker) or
    # All Columns,  and check to see if they share the same marker
    (board[1] == board[4]== board[7] == marker) or
    (board[2] == board[5]== board[8] == marker) or
    (board[3] == board[6]== board[9] == marker) or
    # 2 Diagonals,  and check to see if they share the same marker
    (board[1] == board[5]== board[9] == marker) or
    (board[3] == board[5]== board[7] == marker))


def replay():
    '''Ask player to Play again'''
    choice = input("Play again: Y OR N ").upper()    
    return choice == 'Y'
```

```python
### WHILE LOOP TO KEEP RUNNING THE GAME
print("Welcome to the Tic Tac Toe")

while True:
    
    # play the game
    # SET EVERYTHING UP (BOARD, WHOSE FIRST, CHOOSE MARKER X or O)
    the_board = [' ']*10
    player1_marker, player2_marker = player_input()
    turn = choose_first()
    print(turn + " will go first!")
    
    play_game = input("Ready to Play?! Y or N ").lower()
    if play_game == 'y':
        game_on = True
    else:
        game_on = False
    
    ## THE GAME PLAY
    while game_on:
        
        ### PLAYER 1 TURN 
        if turn == 'Player 1':
            # Show the Board 
            display_board(the_board)
            # choose a position for Player1
            position = player_choice(the_board)
            # place the marker at that position
            place_marker(the_board,player1_marker,position)
            
            # Check if player1 won
            if win_check(the_board,player1_marker):
                display_board(the_board)
                print("Player 1 Has WON!!")
                game_on = False
            else:
                # if board is completely filled then its a Tie
                if full_board_check(the_board):
                    display_board(the_board)
                    print("Tie Game!!")
                    game_on = False
                else:
                    # No Win and NO Tie? Then next player's turn
                    turn = "Player 2"
           
        ### PLAYER 2 TURN 
        else:
            # Show the Board 
            display_board(the_board)
            # choose a position for Player2
            position = player_choice(the_board)
            # place the marker at that position
            place_marker(the_board,player2_marker,position)
            
            # Check if Player2 won
            if win_check(the_board,player2_marker):
                display_board(the_board)
                print("Player 2 Has WON!!")
                game_on = False
            else:
                # Or check for Tie
                if full_board_check(the_board):
                    display_board(the_board)
                    print("Tie Game!!")
                    game_on = False
                else:
                    # No Win and NO Tie? Then next player's turn
                    turn = "Player 1"
    
    if not replay():
        break

# BREAK OUT OF THE WHILE LOOP on replay()
```
